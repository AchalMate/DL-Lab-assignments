```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from tensorflow.keras import Sequential,Model
from tensorflow.python.keras import regularizers
from tensorflow.keras.layers import Dense,Dropout,Activation,MaxPooling2D,Flatten,Conv2D,BatchNormalization,GlobalAveragePooling2D
from tensorflow import keras
from keras.applications.vgg16 import VGG16, preprocess_input
from skimage.transform import resize
```

```python
fastion = tf.keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fastion.load_data()
```

```python
print('X Training shape: ',X_train.shape)
print('Y Training shape: ',y_train.shape)
print('X Testing shape: ',X_test.shape)
print('Y Testing shape: ',y_test.shape)
```

```python
X_train = np.expand_dims(X_train,-1)
X_test = np.expand_dims(X_test,-1)
X_train.shape
```

```python
print(np.unique(y_train))
print(np.unique(y_test))
```

```python
print(X_train.max())
print(X_train.min())
X_train = X_train/255.0
X_test = X_test/255.0
print(X_train.max())
print(X_train.min())
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
def resize_all(x, shape = (48,48)):
    band_shape = x.shape
    x_resize = np.zeros(shape = (band_shape[0],shape[0],shape[1]))
    for i in range(band_shape[0]):
        x_resize[i] = resize(x[i], shape)
    return x_resize


def transform_input_vgg(x):
    x_vgg = np.array(x).reshape(-1,28,28)
    x_vgg = resize_all(x_vgg, (48,48))
    x_vgg = np.repeat(x_vgg[:, :, :, np.newaxis], 3, axis=3)
#    x_vgg = preprocess_input(x_vgg)
    return x_vgg
```

```python
base_model = VGG16(
    weights = "imagenet", 
    input_shape=(48, 48, 3),
    include_top = False
)


for layers in base_model.layers:
    layers.trainable = False


def vgg16_pretrained():
    
    model = Sequential(
        [
            base_model,
            GlobalAveragePooling2D(),
            Flatten(),
            Dense(1024, activation='relu'), 
            Dense(256, activation='relu'),
            Dense(10, activation='softmax')
        ]
    )
    
    return model
```

```python
base_model.summary()
```

```python
model = vgg16_pretrained()
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
X_train = transform_input_vgg(X_train)
X_test = transform_input_vgg(X_test)
```

```python
history = model.fit(X_train,y_train,batch_size=32,epochs=10,validation_split=0.1)
```

```python
print("Training Accuracy")
print(model.evaluate(X_train,y_train))
print("Test accuracy")
print(model.evaluate(X_test,y_test))
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(history.history['loss'], color='b', label="Training Loss")
ax[0].plot(history.history['val_loss'], color='r', label="Validation Loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training Accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation Accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
from tensorflow.keras.applications import ResNet50

pretrained_model=ResNet50( input_shape=(48,48,3),
                                  include_top=False,pooling='max',
                                  weights='imagenet'
                                   )

for layer in pretrained_model.layers:
     layer.trainable = False

pretrained_model.summary()
```

```python
def resnet50_pretrained():
    
    model = Sequential(
        [
            pretrained_model,
            Flatten(),
            Dense(1024, activation='relu'), 
            Dense(256, activation='relu'),
            Dense(10, activation='softmax')
        ]
    )
    
    return model
```

```python
model_resnet = resnet50_pretrained()
model_resnet.summary()
```

```python
model_resnet.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history = model_resnet.fit(X_train,y_train,batch_size=32,epochs=10,validation_split=0.1)
```

```python
print("Training Accuracy")
print(model_resnet.evaluate(X_train,y_train))
print("Test accuracy")
print(model_resnet.evaluate(X_test,y_test))
fig, ax = plt.subplots(2,1)
ax[0].plot(history.history['loss'], color='b', label="Training Loss")
ax[0].plot(history.history['val_loss'], color='r', label="Validation Loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training Accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation Accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```
